```python
from search import search
from IPython.display import display, Image as IPImage
from pathlib import Path
```

```python
def show_results(query: str, top_k: int = 5, use_rewrite: bool = True):
    """搜索 + 展示(文字 + 图片)。"""
    print(f'🔍 查询: "{query}"\n')
    
    results = search(query, top_k=top_k,use_rewrite=use_rewrite)
    
    for r in results:
        print(f"#{r['_rank']}  score={r['_score']:.4f}  id={r['id']}")
        print(f"   character: {r.get('tag_string_character', '-')}")
        print(f"   general  : {r.get('tag_string_general', '')}")
        
        img_path = r.get("local_image_path")
        if img_path and Path(img_path).exists():
            display(IPImage(filename=img_path, width=300))
        else:
            print("   ⚠️ 图片不存在")
        print()
    
    return results
```

```python
results = show_results("龙娘,不要面对镜头的",top_k=10)
results = show_results("角色立绘,不要露出肩膀",top_k=10)
results = show_results("女孩,不要白色背景",top_k=10)
```

```python
from collections import Counter

def analyze_query(query: str, top_k: int = 10, use_rewrite: bool = True):
    """分析查询结果里哪些 tag 出现最多,帮助诊断干扰因素。"""
    results = search(query, top_k=top_k, use_rewrite=use_rewrite)
    
    counter = Counter()
    for r in results:
        tags = r.get('tag_string_general', '').split()
        counter.update(tags)
    
    print(f'🔍 查询 "{query}" 的 Top {top_k} 中,出现最频繁的 tags:\n')
    for tag, count in counter.most_common(20):
        bar = "█" * count
        print(f"  {count}/{top_k}  {bar:<10}  {tag}")
```

```python
analyze_query("A red hair dragon-girl facing the camera", top_k=20)
```
